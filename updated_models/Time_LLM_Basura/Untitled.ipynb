```python
import pandas as pd

# Load the data
df = pd.read_csv('NVDA.csv')  # Replace with the actual path to your CSV file
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler

# Select the columns to scale (excluding 'Date')
columns_to_scale = df.columns.difference(['Date'])

# Initialize the scaler with the desired range
scaler = MinMaxScaler(feature_range=(0, 60))

# Fit and transform the data
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

# Check the scaled data
print(df.head())
```

```python
df.to_csv('NVDA.csv', index=False)  # Replace with the desired output file path
```
